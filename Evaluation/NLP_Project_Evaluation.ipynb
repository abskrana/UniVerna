```python
import os
import json
import warnings
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. SETUP & CONFIGURATION
os.environ["HF_HUB_ENABLE_HF_TRANSFER"] = "1"
os.environ["HF_HUB_DOWNLOAD_TIMEOUT"] = "120"
warnings.filterwarnings('ignore')

from datasets import load_dataset
from sentence_transformers import SentenceTransformer
from FlagEmbedding import BGEM3FlagModel, FlagReranker
from sklearn.preprocessing import MinMaxScaler

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
if DEVICE == "cuda":
    print("GPU:", torch.cuda.get_device_name(0))
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.backends.cudnn.allow_tf32 = True

ALPHA = 0.7
ENCODE_BATCH_SIZE = 64
RERANK_BATCH_SIZE = 32
NUM_QUERIES_TO_TEST = 50

METHOD_NAMES = {
    "Method A": "Baseline Dense (LaBSE)",
    "Method B": "Sparse",
    "Method C": "Dense (BGE)",
    "Method D": "Hybrid",
    "Method E": "Sparse+Rerank",
    "Method F": "Dense+Rerank",
    "Method G": "Hybrid+Rerank",
    "Method H": "Ensemble MRR",
    "Method I": "Ensemble RRF"
}
METHOD_KEYS = list(METHOD_NAMES.keys())

# 2. METRICS & UTILITIES
def calculate_ndcg(ranked, gt, k=10):
    dcg = sum((2**gt.get(d, 0)-1)/np.log2(i+2) for i, d in enumerate(ranked[:k]))
    ideal = sorted(gt.values(), reverse=True)[:k]
    idcg = sum((2**rel-1)/np.log2(i+2) for i, rel in enumerate(ideal))
    return dcg/idcg if idcg > 0 else 0.0

def calculate_recall(ranked, gt, k=5):
    rel = [d for d, r in gt.items() if r > 0]
    return sum(1 for d in ranked[:k] if d in rel)/len(rel) if rel else 0.0

def calculate_mrr(ranked, gt):
    for i, d in enumerate(ranked):
        if gt.get(d, 0) > 0: return 1.0/(i+1)
    return 0.0

def fast_sparse_dot(q_sparse, d_sparse_list):
    return np.array([sum(w * d.get(t, 0.0) for t, w in q_sparse.items()) for d in d_sparse_list])

# 3. LOAD MODELS (ONCE FOR ALL TASKS)
print("\nLoading models...")
baseline_model = SentenceTransformer('sentence-transformers/LaBSE', device=DEVICE)
bge_m3_model = BGEM3FlagModel('BAAI/bge-m3', use_fp16=True, device=DEVICE, use_onnx=False)
reranker_model = FlagReranker('BAAI/bge-reranker-v2-m3', use_fp16=True, device=DEVICE, use_onnx=False)
scaler = MinMaxScaler()

def rerank_top_k(query, corpus, idxs):
    pairs = [[query, corpus[i]] for i in idxs]
    scores = reranker_model.compute_score(pairs, batch_size=RERANK_BATCH_SIZE)
    return [i for i, _ in sorted(zip(idxs, scores), key=lambda x: x[1], reverse=True)]

# 4. CORE SEARCH & EVALUATION PIPELINE
def execute_pipeline(queries, corpus, gt):
    print("Encoding corpus & queries...")
    d_emb_A = baseline_model.encode(corpus, batch_size=ENCODE_BATCH_SIZE, convert_to_tensor=True)
    d_out = bge_m3_model.encode(corpus, batch_size=ENCODE_BATCH_SIZE, return_dense=True, return_sparse=True)
    d_dense, d_sparse = torch.tensor(d_out["dense_vecs"], device=DEVICE), d_out["lexical_weights"]

    q_emb_A = baseline_model.encode(queries, batch_size=ENCODE_BATCH_SIZE, convert_to_tensor=True)
    q_out = bge_m3_model.encode(queries, batch_size=ENCODE_BATCH_SIZE, return_dense=True, return_sparse=True)
    q_dense, q_sparse = torch.tensor(q_out["dense_vecs"], device=DEVICE), q_out["lexical_weights"]

    results = {m: [] for m in METHOD_KEYS}

    print("Searching & Reranking...")
    for i, query in enumerate(queries):
        scores_A = torch.matmul(d_emb_A, q_emb_A[i]).cpu().numpy()
        scores_C = torch.matmul(d_dense, q_dense[i]).cpu().numpy()
        sparse_scores = fast_sparse_dot(q_sparse[i], d_sparse)

        dense_norm = scaler.fit_transform(scores_C.reshape(-1, 1)).flatten()
        sparse_norm = scaler.fit_transform(sparse_scores.reshape(-1, 1)).flatten()
        hybrid = ALPHA * dense_norm + (1 - ALPHA) * sparse_norm

        rankA, rankB, rankC, rankD = (np.argsort(s)[::-1] for s in (scores_A, sparse_scores, scores_C, hybrid))

        results["Method A"].append(rankA)
        results["Method B"].append(rankB)
        results["Method C"].append(rankC)
        results["Method D"].append(rankD)

        results["Method E"].append(rerank_top_k(query, corpus, rankB[:20]))
        results["Method F"].append(rerank_top_k(query, corpus, rankC[:20]))
        results["Method G"].append(rerank_top_k(query, corpus, rankD[:20]))

    print("Computing Ensembles...")
    method_scores = {m: np.mean([calculate_ndcg(results[m][i], gt[i]) for i in range(len(queries))]) for m in METHOD_KEYS[:7]}
    top4_keys = [m[0] for m in sorted(method_scores.items(), key=lambda x: x[1], reverse=True)[:4]]

    for i in range(len(queries)):
        mrr_votes, rrf_votes = {}, {}
        for m in top4_keys:
            for rank, doc_id in enumerate(results[m][i]):
                r = rank + 1
                mrr_votes[doc_id] = mrr_votes.get(doc_id, 0) + (1.0 / r) / 4.0
                rrf_votes[doc_id] = rrf_votes.get(doc_id, 0) + 1.0 / (60 + r)
        results["Method H"].append(sorted(mrr_votes, key=mrr_votes.get, reverse=True))
        results["Method I"].append(sorted(rrf_votes, key=rrf_votes.get, reverse=True))

    metrics_dict = {m: [] for m in METHOD_KEYS}
    for i in range(len(queries)):
        for m in METHOD_KEYS:
            rl = results[m][i]
            metrics_dict[m].append((calculate_ndcg(rl, gt[i]), calculate_recall(rl, gt[i]), calculate_mrr(rl, gt[i])))
            
    return metrics_dict

def display_and_plot(dataset_name, metrics_dict):
    print(f"\n{'='*90}\nFINAL RESULTS: {dataset_name}\n{'='*90}")
    print(f"{'Method':<25} | {'nDCG@10':<10} | {'Recall@5':<10} | {'MRR':<10}")
    print("-" * 90)

    ndcg_vals, recall_vals, mrr_vals = [], [], []
    for m in METHOD_KEYS:
        vals = metrics_dict[m]
        n, r, mr = np.mean([v[0] for v in vals]), np.mean([v[1] for v in vals]), np.mean([v[2] for v in vals])
        ndcg_vals.append(n); recall_vals.append(r); mrr_vals.append(mr)
        print(f"{METHOD_NAMES[m]:<25} | {n:<10.4f} | {r:<10.4f} | {mr:<10.4f}")

    x = np.arange(3)
    width = 0.08
    fig, ax = plt.subplots(figsize=(14, 6))

    for i, m in enumerate(METHOD_KEYS):
        ax.bar(x + (i - 4) * width, [ndcg_vals[i], recall_vals[i], mrr_vals[i]], width, label=METHOD_NAMES[m])

    ax.set_xticks(x); ax.set_xticklabels(["nDCG@10", "Recall@5", "MRR"])
    ax.set_title(f"{dataset_name} Performance Comparison", fontsize=14, fontweight='bold')
    ax.set_ylabel("Score")
    ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# 5. EXECUTE 1: MIRACL DATASET
def fetch_miracl_data(lang_code, num_queries):
    ds = load_dataset("miracl/miracl", lang_code, split="dev", trust_remote_code=True)
    queries, corpus, gt, doc_map = [], [], {}, {}
    for i, item in enumerate(ds.select(range(num_queries))):
        queries.append(item["query"])
        gt_dict = {}
        for p in item["positive_passages"] + item["negative_passages"]:
            text = p["text"]
            if text not in doc_map:
                doc_map[text] = len(corpus)
                corpus.append(text)
            if p in item["positive_passages"]:
                gt_dict[doc_map[text]] = 1
        gt[i] = gt_dict
    return queries, corpus, gt

LANGUAGES = {"en": "English", "hi": "Hindi", "bn": "Bengali", "te": "Telugu"}
miracl_results = {}

for lang_code, lang_name in LANGUAGES.items():
    print(f"\nPROCESSING MIRACL: {lang_name.upper()}")
    mq, mc, mgt = fetch_miracl_data(lang_code, NUM_QUERIES_TO_TEST)
    miracl_results[lang_name] = execute_pipeline(mq, mc, mgt)

# Save to JSON & Load back
with open("results_fast.json", "w") as f:
    json.dump(miracl_results, f, indent=2)

with open("results_fast.json", "r") as f:
    loaded_results = json.load(f)

for lang, metrics in loaded_results.items():
    display_and_plot(f"MIRACL - {lang}", metrics)

# 6. EXECUTE 2: CUSTOM GOVT DATASET
custom_telugu_queries = {
    "q_001": "పుదుచ్చేరి మంత్రిత్వ శాఖలో మత్స్యకారుడు, కుటుంబం, ఆర్థిక సహాయం ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాలు ఏమిటి?",
    "q_002": "విద్య మంత్రిత్వ శాఖలో AICTE, శిక్షణ ట్యాగ్‌లతో ఉన్న పథకం లో అర్హత ఏమిటి?",
    "q_003": "సంబంధిత మంత్రిత్వ శాఖలో అందుబాటు సేవలు ట్యాగ్ ఉన్న పథకం లో వివరాలు ఏమిటి?",
    "q_004": "సూక్ష్మ, చిన్న మరియు మధ్య తరహా పరిశ్రమల మంత్రిత్వ శాఖలో మార్కెటింగ్, సాంకేతికత ట్యాగ్‌లతో ఉన్న పథకం లో వివరాలు ఏమిటి?",
    "q_005": "గుజరాత్ మంత్రిత్వ శాఖలో విద్యార్థి, ఫెలోషిప్ ట్యాగ్‌లతో ఉన్న పథకం లో అర్హత ఏమిటి?",
    "q_006": "ఉత్తరాఖండ్ మంత్రిత్వ శాఖలో విద్య, స్కాలర్‌షిప్ ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాలు ఏమిటి?",
    "q_007": "ఆంధ్రప్రదేశ్ మంత్రిత్వ శాఖలో సామాజిక సంక్షేమం, ఆర్థిక సహాయం ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనం ఏమిటి?",
    "q_008": "కర్ణాటక మంత్రిత్వ శాఖలో వివాహం, ప్రోత్సాహకం ట్యాగ్‌లతో ఉన్న పథకం లో అర్హత ఏమిటి?",
    "q_009": "పుదుచ్చేరి మంత్రిత్వ శాఖలో శిక్షణ, నైపుణ్యం, స్టైపెండ్ ట్యాగ్‌లతో ఉన్న పథకం లో అవసరమైన పత్రాలు ఏమిటి?",
    "q_010": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో పవర్‌లూమ్, ప్రోత్సాహకాలు ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాలు ఏమిటి?",
    "q_011": "రాజస్థాన్ మంత్రిత్వ శాఖలో మహిళ, సాధికారత ట్యాగ్‌లతో ఉన్న పథకం లో అర్హత ఏమిటి?",
    "q_012": "ఉత్తరాఖండ్ మంత్రిత్వ శాఖలో వ్యవసాయం, రైతు, పంట నష్టం ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాలు ఏమిటి?",
    "q_013": "ఛత్తీస్‌గఢ్ మంత్రిత్వ శాఖలో నిర్మాణ కార్మికులు, ఆర్థిక సహాయం ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనం ఏమిటి?",
    "q_014": "రాజస్థాన్ మంత్రిత్వ శాఖలో బీమా, ఆరోగ్యం, కార్మికుడు ట్యాగ్‌లతో ఉన్న పథకం లో అవసరమైన పత్రాలు ఏమిటి?",
    "q_015": "ఉత్తరాఖండ్ మంత్రిత్వ శాఖలో చట్ట అవగాహన, న్యాయ సేవలు ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాలు ఏమిటి?",
    "q_016": "మధ్యప్రదేశ్ మంత్రిత్వ శాఖలో విద్య, వెనుకబడిన వర్గాలు ట్యాగ్‌లతో ఉన్న పథకం లో అర్హత ఏమిటి?",
    "q_017": "ఛత్తీస్‌గఢ్ మంత్రిత్వ శాఖలో కోచింగ్, విద్య ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనం ఏమిటి?",
    "q_018": "కేరళ మంత్రిత్వ శాఖలో మహిళ, రుణం, ఉపాధి ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాలు ఏమిటి?",
    "q_019": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో పర్యాటకం, వ్యాపారం ట్యాగ్‌లతో ఉన్న పథకం లో అర్హత ఏమిటి?",
    "q_020": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో ఉపాధి కల్పన, ప్రోత్సాహకాలు ట్యాగ్‌లతో ఉన్న పథకం లో వివరాలు ఏమిటి?",
    "q_021": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో సామర్థ్య వినియోగం, పర్యాటకం ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాల వివరాలు ఏమిటి?",
    "q_022": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో వడ్డీ రాయితీ, MSME ట్యాగ్‌లతో ఉన్న పథకం లో అర్హత ప్రమాణాలు ఏమిటి?",
    "q_023": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో స్టాంప్ డ్యూటీ, నమోదు రుసుము ట్యాగ్‌లతో ఉన్న పథకం లో దరఖాస్తు విధానం ఏమిటి?",
    "q_024": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో రాష్ట్ర పెట్టుబడి సబ్సిడీ పథకం లో అవసరమైన పత్రాలు ఏమిటి?",
    "q_025": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో నాణ్యత మెరుగుదల, సబ్సిడీ ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాల నిర్మాణం ఏమిటి?",
    "q_026": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో పర్యాటక ప్రోత్సాహం, వడ్డీ రాయితీ ట్యాగ్‌లతో ఉన్న పథకం లో అర్హత ప్రమాణాలు ఏమిటి?",
    "q_027": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో విద్యుత్ సుంకం మినహాయింపు పథకం లో దరఖాస్తు విధానం ఏమిటి?",
    "q_028": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో వస్త్ర పరిశ్రమ, విద్యుత్ పొదుపు ట్యాగ్‌లతో ఉన్న పథకం లో అవసరమైన పత్రాలు ఏమిటి?",
    "q_029": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో వస్త్ర పరిశ్రమ, విద్యుత్ సబ్సిడీ ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాలు ఏమిటి?",
    "q_030": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో వస్త్ర పరిశ్రమ పెట్టుబడి సబ్సిడీ పథకం లో అర్హత ప్రమాణాలు ఏమిటి?",
    "q_031": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో వస్త్ర పరిశ్రమ, జల సంరక్షణ ట్యాగ్‌లతో ఉన్న పథకం లో దరఖాస్తు విధానం ఏమిటి?",
    "q_032": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో వస్త్ర రంగం విద్యుత్ సుంకం మినహాయింపు పథకం లో అవసరమైన పత్రాలు ఏమిటి?",
    "q_033": "పశ్చిమ బెంగాల్ మంత్రిత్వ శాఖలో భూమి మార్పిడి, వస్త్ర పరిశ్రమ ట్యాగ్‌లతో ఉన్న పథకం లో ప్రయోజనాలు ఏమిటి?",
    "q_034": "జమ్మూ కాశ్మీర్ మంత్రిత్వ శాఖలో నిర్మాణ కార్మికుడు, వికలాంగత ట్యాగ్‌లతో ఉన్న పథకం లో అర్హత ప్రమాణాలు ఏమిటి?",
    "q_035": "పుదుచ్చేరి మంత్రిత్వ శాఖలో అధునాతన నైపుణ్య శిక్షణ పథకం లో ప్రయోజనాలు ఏమిటి?",
    "q_036": "పుదుచ్చేరి మంత్రిత్వ శాఖలో కొయిర్ నైపుణ్యాభివృద్ధి లక్ష్యంగా రూపొందించిన శిక్షణ పథకం లో అర్హత ప్రమాణాల నిర్మాణం ఎలా ఉంది?",
    "q_037": "పుదుచ్చేరి మంత్రిత్వ శాఖలో హస్తకళ నైపుణ్యాభివృద్ధి శిక్షణ పథకం లో అందించే ఆర్థిక ప్రయోజనాల స్వభావం ఏమిటి?",
    "q_038": "అస్సాం మంత్రిత్వ శాఖలో టీ పరిశ్రమ వ్యవసాయ ఆదాయపు పన్ను సెలవు పథకం లో ప్రయోజనాల వ్యవధి మరియు స్వరూపం ఏమిటి?",
    "q_039": "పుదుచ్చేరి మంత్రిత్వ శాఖలో నావ యజమానుల బీమా ప్రీమియం సహాయ పథకం లో అర్హత ప్రమాణాలు ఎలా నిర్ణయించబడ్డాయి?",
    "q_040": "పుదుచ్చేరి మంత్రిత్వ శాఖలో పరిశ్రమల ఇంధన మరియు జల సంరక్షణ ఆడిట్ సహాయ పథకం లో అవసరమైన పత్రాలు ఏమిటి?",
    "q_041": "పుదుచ్చేరి మంత్రిత్వ శాఖలో మేధో సంపత్తి హక్కుల రక్షణ సహాయ పథకం లో లభించే ప్రయోజనాల నిర్మాణం ఏమిటి?",
    "q_042": "పుదుచ్చేరి మంత్రిత్వ శాఖలో సాంకేతికత సేకరణ సహాయ పథకం లో అర్హత ప్రమాణాల స్వభావం ఏమిటి?",
    "q_043": "పుదుచ్చేరి మంత్రిత్వ శాఖలో వరి రైతులకు వెనుక-భాగ పెట్టుబడి సబ్సిడీ పథకం లో అర్హత ప్రమాణాలు ఏమిటి?",
    "q_044": "పుదుచ్చేరి మంత్రిత్వ శాఖలో పరిశ్రమల పెట్టుబడి సబ్సిడీ ప్రోత్సాహక పథకం లో దరఖాస్తు విధానం ఎలా ఉంటుంది?",
    "q_045": "పుదుచ్చేరి మంత్రిత్వ శాఖలో పరిశ్రమలకు జనరేటర్ సబ్సిడీ పథకం లో సమర్పించవలసిన పత్రాలు ఏమిటి?",
    "q_046": "పుదుచ్చేరి మంత్రిత్వ శాఖలో హస్తకళ సంఘాలకు ప్రదర్శని గ్రాంటు పథకం లో అందించే ప్రయోజనాల స్వరూపం ఏమిటి?",
    "q_047": "పుదుచ్చేరి మంత్రిత్వ శాఖలో కొయిర్ పరిశ్రమకు మార్జిన్ మనీ రుణ పథకం లో అర్హత ప్రమాణాలు ఏమిటి?",
    "q_048": "పుదుచ్చేరి మంత్రిత్వ శాఖలో పట్టు పరిశ్రమ మూలధన రుణ మార్జిన్ మనీ పథకం లో ప్రయోజనాల నిర్మాణం ఏమిటి?",
    "q_049": "పుదుచ్చేరి మంత్రిత్వ శాఖలో కొయిర్ క్లస్టర్/సంఘాలకు గ్రాంటు పథకం లో సమర్పించవలసిన పత్రాలు ఏమిటి?",
    "q_050": "పుదుచ్చేరి మంత్రిత్వ శాఖలో పట్టు సంఘాలకు గ్రాంటు పథకం లో అర్హత ప్రమాణాల స్వభావం ఏమిటి?"
}

custom_english_queries = {
    "q_001": "What are the benefits of the scheme with the tags Fisherman, Family, and Financial Assistance in the Puducherry Ministry?",
    "q_002": "What is the eligibility for the scheme with AICTE and training tags in the Ministry of Education?",
    "q_003": "What are the details of the scheme with the Accessible Services tag in the relevant ministry?",
    "q_004": "What are the details of the scheme with marketing and technology tags in the Ministry of Micro, Small and Medium Enterprises?",
    "q_005": "What is the eligibility for the scheme with student and fellowship tags in the Gujarat Ministry?",
    "q_006": "What are the benefits of the scheme with education and scholarship tags in the Uttarakhand Ministry?",
    "q_007": "What is the purpose of the scheme with the tags Social Welfare and Financial Assistance in the Ministry of Andhra Pradesh?",
    "q_008": "What is the eligibility for the scheme with marriage and incentive tags in the Karnataka Ministry?",
    "q_009": "What are the documents required in the scheme with training, skill and stipend tags in the Ministry of Puducherry?",
    "q_010": "What are the benefits of the scheme with tags Powerloom and Incentives in the West Bengal Ministry?",
    "q_011": "What is the eligibility for the scheme with women and empowerment tags in the Rajasthan Ministry?",
    "q_012": "What are the benefits of the scheme with tags Agriculture, Farmer, Crop Loss in the Uttarakhand Ministry?",
    "q_013": "What is the purpose of the scheme with tags for construction workers and financial assistance in the Chhattisgarh Ministry?",
    "q_014": "What are the documents required in the scheme with insurance, health, and worker tags in the Rajasthan Ministry?",
    "q_015": "What are the benefits of the scheme with tags Legal Awareness and Legal Services in the Uttarakhand Ministry?",
    "q_016": "What is the eligibility for the scheme with tags Education and Backward Classes in the Ministry of Madhya Pradesh?",
    "q_017": "What is the purpose of the scheme with coaching and education tags in the Chhattisgarh Ministry?",
    "q_018": "What are the benefits of the scheme with the tags of women, loan and employment in the Kerala Ministry?",
    "q_019": "What is the eligibility for the scheme with tourism and business tags in the West Bengal Ministry?",
    "q_020": "What are the details of the scheme with tags Employment Generation and Incentives in the West Bengal Ministry?",
    "q_021": "What are the details of the benefits of the scheme with tags Capacity Utilization and Tourism in the West Bengal Ministry?",
    "q_022": "What are the eligibility criteria for the scheme with interest subsidy and MSME tags in the West Bengal Ministry?",
    "q_023": "What is the application procedure for the scheme with stamp duty and registration fee tags in the West Bengal Ministry?",
    "q_024": "What are the documents required for the State Investment Subsidy Scheme in the West Bengal Ministry?",
    "q_025": "What is the benefit structure in the scheme with quality improvement and subsidy tags in the West Bengal Ministry?",
    "q_026": "What are the eligibility criteria for the scheme with tourism promotion and interest subsidy tags in the West Bengal Ministry?",
    "q_027": "What is the application procedure for the Electricity Tariff Exemption Scheme in the West Bengal Ministry?",
    "q_028": "What are the documents required in the scheme with energy saving tags for the textile industry in the West Bengal Ministry?",
    "q_029": "What are the benefits of the scheme with textile industry and electricity subsidy tags in the West Bengal Ministry?",
    "q_030": "What are the eligibility criteria for the Textile Industry Investment Subsidy Scheme in the West Bengal Ministry?",
    "q_031": "What is the application procedure for the scheme with textile industry and water conservation tags in the West Bengal Ministry?",
    "q_032": "What are the documents required for the Textile Sector Electricity Tariff Exemption Scheme in the West Bengal Ministry?",
    "q_033": "What are the benefits of the scheme with land exchange and textile industry tags in the West Bengal Ministry?",
    "q_034": "What are the eligibility criteria for the scheme under the Jammu and Kashmir Ministry for Construction Workers and Disabled with tags?",
    "q_035": "What are the benefits of the Advanced Skill Training Scheme in the Ministry of Education, Puducherry?",
    "q_036": "How are the eligibility criteria structured in the training scheme designed for the purpose of developing coir skills in the Ministry of Puducherry?",
    "q_037": "What is the nature of financial benefits provided under the Handicraft Skill Development Training Scheme in the Ministry of Puducherry?",
    "q_038": "What is the duration and nature of benefits under the Tea Industry Agricultural Income Tax Holiday Scheme in the Ministry of Assam?",
    "q_039": "How are the eligibility criteria determined in the Ship Owners Insurance Premium Assistance Scheme in the Puducherry Ministry?",
    "q_040": "What are the documents required under the Industrial Energy and Water Conservation Audit Assistance Scheme in the Ministry of Puducherry?",
    "q_041": "What is the structure of benefits available under the Intellectual Property Rights Protection Assistance Scheme in the Ministry of Puducherry?",
    "q_042": "What is the nature of eligibility criteria in the Technology Acquisition Assistance Scheme in the Ministry of Puducherry?",
    "q_043": "What are the eligibility criteria for the Backward-Investment Subsidy Scheme for Paddy Farmers in the Puducherry Ministry?",
    "q_044": "What is the application procedure for the Industrial Investment Subsidy Promotion Scheme in the Ministry of Puducherry?",
    "q_045": "What are the documents required to be submitted under the Generator Subsidy Scheme for Industries in the Puducherry Ministry?",
    "q_046": "What are the benefits provided under the Demonstration Grant Scheme to Handicraft Societies in the Ministry of Puducherry?",
    "q_047": "What are the eligibility criteria for the Margin Money Loan Scheme for Coir Industry in the Puducherry Ministry?",
    "q_048": "What is the benefit structure in the Silk Industry Capital Loan Margin Money Scheme in the Puducherry Ministry?",
    "q_049": "What are the documents required to be submitted under the grant scheme for Choir Clusters/Associations in the Ministry of Puducherry?",
    "q_050": "What is the nature of the eligibility criteria for the grant scheme for silk societies in the Puducherry Ministry?"
}

custom_qrels = {
    "q_001": {"doc_001_benefits": 2},"q_002": {"doc_002_eligibility": 2}, "q_003": {"doc_003_details": 2},
    "q_004": {"doc_004_details": 2}, "q_005": {"doc_005_eligibility": 2}, "q_006": {"doc_006_benefits": 2},
    "q_007": {"doc_007_benefits": 2}, "q_008": {"doc_008_eligibility": 2}, "q_009": {"doc_009_documents_required": 2},
    "q_010": {"doc_010_benefits": 2}, "q_011": {"doc_011_eligibility": 2}, "q_012": {"doc_012_benefits": 2},
    "q_013": {"doc_013_benefits": 2}, "q_014": {"doc_014_documents_required": 2}, "q_015": {"doc_015_benefits": 2},
    "q_016": {"doc_016_eligibility": 2}, "q_017": {"doc_017_benefits": 2}, "q_018": {"doc_018_benefits": 2},
    "q_019": {"doc_019_eligibility": 2}, "q_020": {"doc_020_details": 2}, "q_021": {"doc_021_benefits": 2},
    "q_022": {"doc_022_eligibility": 2}, "q_023": {"doc_023_application_process": 2}, "q_024": {"doc_024_documents_required": 2},
    "q_025": {"doc_025_benefits": 2}, "q_026": {"doc_026_eligibility": 2}, "q_027": {"doc_027_application_process": 2},
    "q_028": {"doc_028_documents_required": 2}, "q_029": {"doc_029_benefits": 2}, "q_030": {"doc_030_eligibility": 2},
    "q_031": {"doc_031_application_process": 2}, "q_032": {"doc_032_documents_required": 2}, "q_033": {"doc_033_benefits": 2},
    "q_034": {"doc_034_eligibility": 2}, "q_035": {"doc_035_benefits": 2}, "q_036": {"doc_036_eligibility": 2},
    "q_037": {"doc_037_benefits": 2}, "q_038": {"doc_038_benefits": 2}, "q_039": {"doc_039_eligibility": 2},
    "q_040": {"doc_040_documents_required": 2}, "q_041": {"doc_041_benefits": 2}, "q_042": {"doc_042_eligibility": 2},
    "q_043": {"doc_043_eligibility": 2}, "q_044": {"doc_044_application_process": 2}, "q_045": {"doc_045_documents_required": 2},
    "q_046": {"doc_046_benefits": 2}, "q_047": {"doc_047_eligibility": 2}, "q_048": {"doc_048_benefits": 2},
    "q_049": {"doc_049_documents_required": 2}, "q_050": {"doc_050_eligibility": 2}
}

print("\nPROCESSING CUSTOM GOV CORPUS")
try:
    with open('gov_corpus.json', 'r', encoding='utf-8') as f:
        hierarchical_corpus = json.load(f)

    corpus_ids, corpus_texts = [], []
    expected_sections = ["Details", "Benefits", "Eligibility", "Application Process", "Documents Required"]

    print("Building semantic chunks")
    for doc_id, doc_data in hierarchical_corpus.items():
        ministry = doc_data.get("Ministry", "Unknown")
        title = doc_data.get("Title", "Unknown")
        tags = ", ".join(doc_data.get("Tags", []))

        for section in expected_sections:
            if section in doc_data:
                chunk_id = f"{doc_id}_{section.lower().replace(' ', '_')}"
                chunk_text = (f"Ministry: {ministry}\nScheme Title: {title}\n"
                              f"Tags: {tags}\nSection: {section}\nInformation: {doc_data[section]}")
                corpus_ids.append(chunk_id)
                corpus_texts.append(chunk_text)
    

    for language, custom_queries in [["Telugu", custom_telugu_queries], ["English", custom_english_queries]]:
        doc_index = {doc_id: i for i, doc_id in enumerate(corpus_ids)}
        query_texts = [custom_queries[q] for q in custom_queries.keys()]

        ground_truth = {}
        for i, qid in enumerate(custom_queries.keys()):
            ground_truth[i] = {doc_index[doc_id]: rel for doc_id, rel in custom_qrels[qid].items() if doc_id in doc_index}

        gov_metrics = execute_pipeline(query_texts, corpus_texts, ground_truth)
        display_and_plot(f"Custom Govt Dataset ({language})", gov_metrics)

except FileNotFoundError:
    print("Skipping Custom Govt Dataset: 'gov_corpus.json' file not found.")
```
